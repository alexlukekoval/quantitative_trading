```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import glob
import seaborn as sns
sns.set()
```

```python
file = glob.glob('*.ftr')[0]
print(f'file: {file}')
```

```python
df = pd.read_feather(file)
```

```python
df.shape
```

```python
df.columns
```

```python
df_subset = df[:60]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_subset['date'], df_subset['ask'], label='Ask Price')
ax.plot(df_subset['date'], df_subset['bid'], label='Bid Price')
ax.plot(df_subset['date'], df_subset['lastPrice'], label='Last Price', marker='o', markersize=3, linestyle='--', color='black', alpha=0.5)

ax.set_xlabel('Time')
ax.set_ylabel('BTCUSD~coinbase')
# Set y-axis tick label format to 'plain' and turn off the offset
ax.ticklabel_format(axis='y', style='plain', useOffset=False)
ax.set_title('Plot to show bid-ask bounce')
ax.legend()
plt.show()


```
